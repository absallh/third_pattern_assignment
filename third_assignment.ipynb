# Start preprocessing the data 

```python
#extract feature
#add the length of the massage as feature
data["Length"] = data["message"].apply(len)
```

```python
#make the spam represented by 1 and ham represented by 0
data['label']=data['label'].map({'ham':0, 'spam':1})
```

```python
#split the data to x and y
# x = length and y = the label
x = data['Length'].values
# y = 1 if spam 0 if ham
y = np.array([(1 if label == 'spam' else 0) for label in data['label'].values])
y
```

```python
#split the data to train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7)
```

# Some helper functions

```python
def make_random_w ():
    #create random Weights and bias
    np.random.seed(1)
    w = np.random.rand()    
    b = np.random.rand()
    return w, b
#call the function
make_random_w()
```

```python
def f(z):
    #predict function
    output = 1 / (1 + np.exp(-z))
    return output
```

```python
def cost (y, t):
    #crose entorpy function
    #if t = 1 then E = log(y)
    #if t = 0 then E = log(1-y)
    return -(1/len(t)) * np.sum(t * np.log(y) + (1 - t) * np.log(1-y))
```

# Train the model with normal equation

```python
def gd(x, t): 
    size = len(t)
    lr = 0.001
    weight, bias = make_random_w()
    y = f(np.dot(x, weight) + bias)
    current_loss = cost (y, t)
    while True:        
        dW = 1/size * np.dot(x.T, (y - t))
        db = 1/size * np.sum(y - t)
        weight -= lr * dW
        bias -= lr * db
        old_loss = current_loss
        y = f(np.dot(x, weight) + bias)
        current_loss = cost (y, t)
        print(current_loss)
        if (old_loss - current_loss) < 0.000000005 : break
    
    return [weight, bias]
```

```python
weight, bias = gd(x_train, y_train)
```

```python
print(weight, "\t", bias)
```

# calculate the accuracy

```python
def accuracy (y, t):
    corect = 0
    for i in range(len(t)):
        if t[i] == round(y[i]):
            corect += 1
    return (corect/len(t))
```

```python
y = f(np.dot(x_test, weight) + bias)
accuracy(y, y_test)
```
